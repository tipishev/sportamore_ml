```python
print(__doc__)


# Code source: GaÃ«l Varoquaux
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


from sklearn.cluster import KMeans
from sklearn import datasets

np.random.seed(5)

centers = [[1, 1], [-1, -1], [1, -1]]
iris = datasets.load_iris()
X = iris.data
y = iris.target

estimator = KMeans(n_clusters=3, n_init=1,  max_iter=2, 
                   algorithm='full', init='random')

fig = plt.figure(figsize=(1, 0.75))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, 9.5, 10], elev=48, azim=134)

plt.cla()
estimator.fit(X)
labels = estimator.labels_

ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=labels.astype(np.float))
cc = estimator.cluster_centers_
# TODO color these so we can see them with the above scatter
ax.scatter(cc[:, 3], cc[:, 0], cc[:, 2], s=400)


ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')

plt.show()
```
